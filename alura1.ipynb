```python
import numpy as np

#criao de um daframe de teste para usar de exemplo
df_salarios = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Vanessa", "Richard", "val"],
       'salarios': [4000, np.nan, 5000, np.nan, 10000]
})
# calcula a media salarial e substitui os nulos pela media e arredonada os valores
df_salarios['salarios_media'] = df_salarios['salarios'].fillna(df_salarios['salarios'].mean().round(2))

''' calcula a media e substitui os nulos pela mediana e arredondada os valores'''

df_salarios['salarios_mediana'] = df_salarios['salarios'].fillna(df_salarios['salarios'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["segunda", "tera", "Quarta", "Quinta", "Sexta"],
     "Temperatura": [30, np.nan, np.nan, 28, 27]

  })

df_temperaturas["preenchido_ffill"]= df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["segunda", "tera", "Quarta", "Quinta", "Sexta"],
     "Temperatura": [30, np.nan, np.nan, 28, 27]

  })

df_temperaturas["preenchido_bfill"]= df_temperaturas["Temperatura"].bfill()
df_temperaturas
```

```python
import pandas as pd
import numpy as np

df_vendas = pd.DataFrame({
    "produtos": [ "roupa", "sapato", np.nan, "tenis", "chilelo"],
    "preco": [10.20, np.nan, 15.00, np.nan, np.nan],
    "quantidade": [np.nan, np.nan, np.nan, np.nan, np.nan],
})

mean_preco = df_vendas['preco'].mean()

if pd.isna(mean_preco):
  df_vendas['vendas_media_preco'] = df_vendas['preco'].fillna(0)
  df_vendas['vendas_media_quantidade'] = df_vendas['quantidade'].fillna(0)
else:
  df_vendas['vendas_media_preco'] = df_vendas['preco'].fillna(mean_preco.round(2))
  df_vendas['vendas_media_quantidade'] = df_vendas['quantidade'].fillna(mean_preco.round(2))

df_vendas
```
